```python
#importing required modules
import pandas as pd
import numpy as np

#opening  the file with open method and transferring the data to  a string
data = open("SMSSpamCollection.txt","r")

string1=""
while data.readline():
  string1+=data.readline()
print(type(string1))
```

```python
#Splitting the data into words and sentences and storing the word count
words_data=string1.split(" ")
print(words_data)
words_count=len(words_data)
sentence_data=list(string1.split("."))
print(len(sentence_data),words_count)
```

```python
#Dividing the data into sentence 
punctuation = [",", ".", ":", ";", "!", "?"]
for i, sentence in enumerate(sentence_data.copy()):
    new_sentence = [word for word in sentence if word not in punctuation]
    sentence_data[i] = new_sentence
```

```python
for i, sentence in enumerate(sentence_data.copy()):
    new_sentence = list()
    for j, word in enumerate(sentence.copy()):
        new_word = word.lower() # Lower case all characters in word
        new_sentence.append(new_word)
    sentence_data[i] = new_sentence
#print(new_sentence[1:10])
```

```python
import nltk
new_words = list()
for sentence in sentence_data:
    for word in sentence:
        new_words.append(word)
unigram_fdist = nltk.FreqDist(new_words)
```

```python
def getUnigramProbability(word):
    if word in unigram_fdist:
        return unigram_fdist[word]/words_count
    else:
        return -1
```

```python
bigrams = list()
for sentence in sentence_data:
    new_bigrams = nltk.bigrams(sentence)
    bigrams += new_bigrams
```

```python
bigram_fdist = nltk.ConditionalFreqDist(bigrams)
def getBigramProbability(word1, word2):
    if word1 not in bigram_fdist:
        return -1 
    elif word2 not in bigram_fdist[word1]:
        return getUnigramProbability(word2)
    else:
        bigram_frequency = bigram_fdist[word1][word2]
        unigram_frequency = unigram_fdist[word1]
        bigram_probability = bigram_frequency / unigram_frequency
        return bigram_probability
```
